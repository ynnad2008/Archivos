```r
#Directorio de trabajo
#directorio <- "/home"
imagen <- "Extract_DEM.tif"

#Lista de paquetes de Interes
packs <- c('raster')

#for (pkg in packs) {
#  if (!require(pkg, character.only = TRUE)) {
#    install.packages(pkg, dependencies = TRUE)
#  }
#}

library(raster)

print("Directorio de trabajo")
getwd()

ls()

r <- raster(imagen)
r

dim(r)
plot(r, col=rainbow(50), alpha=0.37, main='Altimetria')

slp <- terrain(r, opt="slope", unit="degrees", filename="pendientes.tif", overwrite=TRUE)
pslp<-plot(slp)
title("Modelo de Pendientes")

rgdns <- terrain(r, opt="roughness", filename="rugosidad.tif", overwrite=TRUE)
prgdns<-plot(rgdns)
title("Rugosidad")

aspct <- terrain(r, opt="aspect", filename="aspecto.tif", overwrite=TRUE)
paspct<-plot(aspct)
title("Aspecto")

tpi <- terrain(r, opt="TPI", filename="TPI.tif", overwrite=TRUE)
ptpi<-plot(tpi)
title("Indice TPI")

tri <- terrain(r, opt="TRI", filename="TRI.tif", overwrite=TRUE)
ptri<-plot(tri)
title("Indice TRI")

fdir <- terrain(r, opt="flowdir", filename="flowdir.tif", overwrite=TRUE)
pfdir<-plot(fdir)
title("Dirección de Flujo")

####
png(filename = "FlowDir110.png", res = 110)
plot(fdir, main="Dirección de Flujo")
dev.off()
###

par(mfrow = c(2, 3))
plot(slp)
title("Modelo de Pendientes")
plot(aspct)
title("Aspecto")
plot(rgdns)
title("Rugosidad")
plot(fdir)
title("Dirección de Flujo")
plot(tpi)
title("Indice TPI")
ptri<-plot(tri)
plot(r, col=rainbow(25, alpha=0.35), legend=FALSE, add=TRUE)
title("Indice Rugosidad Terreno")

hs <- hillShade(terrain(r, opt="slope", unit="radians"),
          terrain(r, opt="aspect",unit="radians"),
          angle=45, direction=0, filename='hillshade.tif', normalize=FALSE, overwrite=TRUE)
plot(hs, col=grey(0:100/100), legend=FALSE)


plot(hs, col=grey(0:100/100), legend=FALSE, main='Area de estudio')
plot(r, col=rainbow(25, alpha=0.35), add=TRUE)
```
